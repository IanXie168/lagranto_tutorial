```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pylab as plt
%matplotlib inline  
from dypy.plotting import Mapfigure
from mpl_toolkits.basemap import Basemap
```

```python
from dypy.lagranto import Tra
trajs = Tra()
```

```python
trajs.load_ascii('lagrantodata/wcb.1')

variables = ['p', 'Q']
titles = ['Elevation, hPa', 'Specific humidity, g/kg']

for i in range(2):
    figname = 'lagrantodata/' + variables[i] + '.png'


    fig = plt.figure(figsize = (10,8))
    m = Mapfigure(width=6500000,height=5500000,
            resolution='l',projection='laea',\
            lat_ts=60,lat_0=62,lon_0=10.)
    m.drawmap(nbrem=10, nbrep=10)    
    cont = m.plot_traj(trajs, variables[i], cmap = 'jet')

    cb = fig.colorbar(cont)    
    for font_objects in cb.ax.yaxis.get_ticklabels():
        font_objects.set_size(20)
    plt.title(titles[i], fontsize=24)
    
    fig.savefig(figname, bbox_inches='tight')
```
